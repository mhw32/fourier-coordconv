```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
import torchvision.utils
```

```python
timestep = 100
perturbed_multimnist_vanilla_loss = np.load('/mnt/fs5/wumike/fourier/perturbed_mnist/vanilla/loss.npy')[:timestep, 1]
perturbed_multimnist_vanilla_acc = np.load('/mnt/fs5/wumike/fourier/perturbed_mnist/vanilla/accuracy.npy')[:timestep, 1]
perturbed_multimnist_coord_loss = np.load('/mnt/fs5/wumike/fourier/perturbed_mnist/coord/loss.npy')[:timestep, 1]
perturbed_multimnist_coord_acc = np.load('/mnt/fs5/wumike/fourier/perturbed_mnist/coord/accuracy.npy')[:timestep, 1]
perturbed_multimnist_addfourier_loss = np.load('/mnt/fs5/wumike/fourier/perturbed_mnist/add_fourier/loss.npy')[:timestep, 1]
perturbed_multimnist_addfourier_acc = np.load('/mnt/fs5/wumike/fourier/perturbed_mnist/add_fourier/accuracy.npy')[:timestep, 1]
perturbed_multimnist_catfourier_loss = np.load('/mnt/fs5/wumike/fourier/perturbed_mnist/cat_fourier/loss.npy')[:timestep, 1]
perturbed_multimnist_catfourier_acc = np.load('/mnt/fs5/wumike/fourier/perturbed_mnist/cat_fourier/accuracy.npy')[:timestep, 1]
```

```python
plt.figure(figsize=(10,10))
plt.plot(perturbed_multimnist_vanilla_acc, label='Conv')
plt.plot(perturbed_multimnist_coord_acc, label='CoordConv')
plt.plot(perturbed_multimnist_addfourier_acc, label='FourierCoordConv [Add]')
plt.plot(perturbed_multimnist_catfourier_acc, label='FourierCoordConv [Cat]')
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.legend(fontsize=10)
plt.title('Perturbed MNIST', fontsize=16)
plt.show()
```
